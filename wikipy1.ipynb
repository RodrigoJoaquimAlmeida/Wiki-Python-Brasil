João Papo-de-Pescador, homem de bem, comprou um microcomputador para controlar o rendimento diário de seu trabalho. Toda vez que ele traz um peso de peixes maior que o estabelecido pelo regulamento de pesca do estado de São Paulo (50 quilos) deve pagar uma multa de R$ 4,00 por quilo excedente. João precisa que você faça um programa que leia a variável peso (peso de peixes) e calcule o excesso. Gravar na variável excesso a quantidade de quilos além do limite e na variável multa o valor da multa que João deverá pagar. Imprima os dados do programa com as mensagens adequadas.

```python
reg_sp = 50
multa = 4

peixe = int(input('Digite o peso do peixe: '))

if peixe > reg_sp:
    excesso = peixe - reg_sp
    valor_m = excesso * multa
    print('Joao excedeu a quantidade em {} e pagara uma multa de {} reais'.format(excesso, valor_m))
```

Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês. Calcule e mostre o total do seu salário no referido mês, sabendo-se que são descontados 11% para o Imposto de Renda, 8% para o INSS e 5% para o sindicato, faça um programa que nos dê:<br>
salário bruto.<br>
quanto pagou ao INSS.<br>
quanto pagou ao sindicato.<br>
o salário líquido.<br>
calcule os descontos e o salário líquido, conforme a tabela abaixo:<br>
'valor Salário Bruto em reais' <br>
'menos IR (11 %) em reais' <br>
'menos INSS (8 %) em reais' <br>
'menos Sindicato ( 5 %) em reais' <br>
'igual Salário Liquido em reais' <br>
Obs.: Salário Bruto - Descontos = Salário Líquido.

```python
valor_hora = float(input('Digite seu valor/hora: '))
hora_trab = int(input('Digite o total de horas no mes: '))

salario_brut = valor_hora * hora_trab
inss = salario_brut * 0.08
sindic = salario_brut * 0.05


salario_liq = salario_brut - salario_brut * 0.11 - inss - sindic

print('O salaro bruto no mês foi de R$ {} sendo INSS no valor de R$ {} o sindicato R$ {} e o líquido de R$ {}'.format(salario_brut, inss, sindic, salario_liq))
```

Faça um programa para uma loja de tintas. O programa deverá pedir o tamanho em metros quadrados da área a ser pintada. Considere que a cobertura da tinta é de 1 litro para cada 3 metros quadrados e que a tinta é vendida em latas de 18 litros, que custam R$ 80,00. Informe ao usuário a quantidades de latas de tinta a serem compradas e o preço total.

```python
import math
area_pint = float(input('Digite a area em m² a ser pintada: '))
preco_lata = 80
litros_gastos = area_pint / 3.00
lata_total = math.ceil(litros_gastos / 18.00)
custo = math.ceil(lata_total * preco_lata)

print('O total de latas é de {} unidade, e o valor gasto é de R$ {}'.format(lata_total, custo))
```

Informe ao usuário as quantidades de tinta a serem compradas e os respectivos preços em 3 situações:<br>
comprar apenas latas de 18 litros;<br>
comprar apenas galões de 3,6 litros;<br>
misturar latas e galões, de forma que o desperdício de tinta seja menor. Acrescente 10% de folga e sempre arredonde os valores para cima, isto é, considere latas cheias.

```python
import math

area_pintada = int(input('Digite a area a ser pintada, em m²: '))


cobertura_1ltinta = area_pintada / 6

if cobertura_1ltinta % 18 == 0:
    lata = math.ceil(cobertura_1ltinta / 18)
    preco1 = lata * 80
    print('Total de lata e {} e o custo de R$ {}'.format(lata, preco1))
    
elif cobertura_1ltinta % 3.6 == 0:
    galao = math.ceil(cobertura_1ltinta / 3.6)
    preco2 = galao * 25
    print('Total de galoes e {} e o custo de R$ {}'.format(galao, preco2))
    
else:
    cobertura_1ltinta *= 1.1
    lata = math.floor(cobertura_1ltinta / 18)
    faltando_litros = cobertura_1ltinta - (lata * 18)
    galao = math.ceil(faltando_litros / 3.6)
        
    if galao <= 3:
        preco2 = galao * 25
        preco1 = lata * 80
        print('Total de galoes e {} total de latas e {} o custo de galoes e R$ {} e o custo de latas e R$ {}'.format(galao, lata, preco2, preco1))
    else:
        lata += 1
        preco1 = lata * 80
        galao = 0
        preco2 = galao * 25
        print('Total de lata e {} e o total de galoes e {} e o custo lata e de R$ {} e o custo galoes e de R$ {}'.format(lata, galao, preco1, preco2))
        
    


    
```
